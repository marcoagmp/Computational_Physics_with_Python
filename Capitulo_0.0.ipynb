#### A função linspace() abaixo é muito útil para testar valores. Ao invés de precisar **digitar** números para todos os pontos da ordenada *time*, a gente simplesmente fala pro Python que queremos um espaçamento linear de números entre 0.0 a 10.0 com 100 números no total. Isso é muito útil para testar valores das funções e enxergar como determinada relação se comporta.
#### Nota que o argumento que vem depois das coordenadas ('r-' e 'g-') indicam a cor do plot! Equanto o elemento '^' que vem acompanhado no 'r-' indica ao plot a vontade de colocar no gráfico cada ponto chamado na função linspace().

```python
time = linspace(0.0, 10.0, 100)
height = exp(-time/3.0)*sin(time*3)
plot(time, height, 'r-^')
plot(time, 0.3*sin(time*3), 'b-')
legend(['damped', 'constant amplitude'], loc='upper right') #loc é a localização da legenda no gráfico
xlabel('Time (s)') #legenda do eixo x
```

#### 0.3 - Problemas:
##### 0-0 Coloque em um gráfico as funções abaixo em uma única figura, utilizando uma escala apropriada:
$$a)\ x^4e^{-2x}$$
$$b)\ [x²e^{-x}sin(x²)]²$$


```python
#Você pode ir testando o argumento '200' para ver como a forma do plot vai se alterando
#É interessante observar a necessidade de se criar um numpy.array, de forma que o cálculo do eixo_y é realizado para cada índice do np.array
eixo_x = np.linspace(0,7,200)
eixo_y = eixo_x**4*np.exp(-2*eixo_x)
eixo_y2 = (eixo_x**2 * np.exp(-eixo_x) * np.sin(eixo_x**2))**2
plot(eixo_x,eixo_y, 'm')
plot(eixo_x,eixo_y2, 'b')
legend(['a','b'], loc='upper right')
xlabel('Eixo x')
ylabel('Eixo y')
```
