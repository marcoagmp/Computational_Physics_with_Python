#### A função linspace() abaixo é muito útil para testar valores. Ao invés de precisar **digitar** números para todos os pontos da ordenada *time*, a gente simplesmente fala pro Python que queremos um espaçamento linear de números entre 0.0 a 10.0 com 100 números no total. Isso é muito útil para testar valores das funções e enxergar como determinada relação se comporta.
#### Nota que o argumento que vem depois das coordenadas ('r-' e 'g-') indicam a cor do plot! Equanto o elemento '^' que vem acompanhado no 'r-' indica ao plot a vontade de colocar no gráfico cada ponto chamado na função linspace().

```python
time = np.linspace(0.0, 10.0, 100)
height = np.exp(-time/3.0)*np.sin(time*3)
plt.plot(time, height, 'r-^')
plt.plot(time, 0.3*np.sin(time*3), 'b-')
plt.legend(['damped', 'constant amplitude'], loc='upper right') #loc é a localização da legenda no gráfico
plt.xlabel('Time (s)') #legenda do eixo x
plt.show()
```

##### 0-3 O arquivo 'Ba137.txt' na verdade contém dados de um experimento de decaimento radioativo; a primeira coluna (Contador Geiger) é o número de decaimentos _N_ e a segunda coluna é o tempo _t_ em segundos. É interessante saber o tempo de meia vida $t_{(1/2)}$ da espécie química $^{137}\text{Ba}$. Ela segue a equação de decaimento $$N = N_o\exp^{-\lambda t}$$ onde $\lambda = \frac{\log(2)}{t_{1/2}}$. Utilizando técnicas que você aprendeu no capítulo, carregue o arquivo do Ba137.txt em variáveis apropriadas e plote a equação resultante. Qual a melhor estimativa para o tempo de meia vida $t_{(1/2)}$?

##### Para encontrar a melhor estimativa para o tempo de meia vida, primeiro é preciso linearizar a função $N = N_o\exp^{-\lambda t}$ de forma a encontrar uma estimativa da melhor inclinação da reta e do intercepto/coeficiente linear (ponto da reta que intercepta o eixo y). Para fazer isso, aplicamos o logaritmo natural  nos dois lados da equação $\ln(N) = \ln(N_o e^{-\lambda t})$ e aplicando propriedade dos logaritmos $\ln(ab) = \ln(a) + \ln(b)$, a equação se torna: $$\ln(N) = \ln(N_o) + \ln(e^{-\lambda t})$$ Podemos simplificá-la para $$\ln(N) = \ln(N_o) - \lambda t$$

##### Esta é uma equação linear do tipo $y = c + mx$, onde:
- $y = \ln(N)$
- $x = t$
- $c(intercepto) = \ln(N_o)$
- $m = -\lambda$ (inclinação da reta/coeficiente angular)

##### Conforme a equação linear, o intercepto da reta é igual a $\ln(N_o)$. Para encontrar $N_o$, basta reverter a operação, aplicando a função exponencial ($e^{x}$): $$N_o = e^{c}$$

##### Com a equação linear, foi possível usar a regressão linear (utilizando o tempo _t_ como variável independente e $\ln(N)$ como variável dependente), para encontrar a reta que melhor se ajusta a um conjunto de dados. O resultado desse ajuste é uma variável que fornece a inclinação (_m_) e o valor do intercepto/coeficiente linear (_c_). 

```python
import pandas as pd
from scipy.stats import linregress #Função principal para a regressão linear. Ela encontra a reta que melhor se ajusta aos dados e nos fornece a inclinação e o intercepto.
```

```python
#names=['N', 't']: Atribui nomes às colunas para facilitar a manipulação. A primeira coluna (N) é a contagem, e a segunda (t) é o tempo.

#decimal='.': Garante que o Python interprete o ponto (.) como o separador decimal, o que é comum em arquivos de dados em inglês.
data = pd.read_csv('Ba137.txt', sep='\t', header=None, names=['N', 't'], decimal='.')
```

```python
#Linearização da equação:
data['log_N'] = np.log(data['N'])
```

```python
#Realização da regressão linear:
#linregress(data['t'], data['log_N']): Esta é a chamada da função que faz a mágica. Ela recebe a coluna de tempo (t) como eixo x e a nova coluna de logaritmo (log_N) como eixo y.

#A função linregress retorna um objeto com os resultados do ajuste.

#lin_reg.intercept: Atribui o valor do intercepto (ponto onde a linha cruza o eixo y) à variável intercept. Conforme a nossa equação linearizada, este valor é igual a ln(N_o).

#lin_reg.slope: Atribui a inclinação da reta à variável slope. Este valor é igual a −lambda.

lin_reg = linregress(data['t'], data['log_N'])
intercept = lin_reg.intercept
slope = lin_reg.slope
print(intercept)
print(slope)
```

```python
#Cálculo de N_o
N_o = np.exp(intercept)
print(N_o)
```

```python
#Cálculo do tempo de meia vida (t_(1/2)):
tempo_meia_vida = np.log(2) / -(slope)
print(tempo_meia_vida)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['t'], data['log_N'], label='Dados originais (log(N) vs t)', color='blue')
plt.plot(data['t'], intercept + slope * data['t'], color='red', label=f'Ajuste linear: log(N) = {intercept:.2f} + ({slope:.4f})t')
plt.title('Ajuste Linear para Encontrar N_o e Lambda')
plt.xlabel('Tempo (t) em segundos')
plt.ylabel('log(N)')
plt.legend()
plt.grid(True)
plt.show()
```

##### Calculando o tempo de meia vida $t_{1/2}$ a partir dos dados, fazemos uma estimativa de que ele é aproximadamente $\approx$ 167.29s, que se aproxima bem do valor da literatura...
