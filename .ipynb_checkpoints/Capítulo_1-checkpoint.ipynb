## 1.11- Problemas

### 1-2 Um projétil lançado verticalmente em uma velocidade inicial $v_i$ tem a posição dada pela fórmula $y(t) = y_i + v_it - \frac{1}{2}gt²$, onde g = 9.8 m/s². Escreva um código Python que cria duas listas: uma contendo dados do tempo (50 pontos em 5 segundos) e a outra contendo os dados correspondentes a esses valores de acordo com a fórmula acima. O programa deve perguntar pela altura inicial ($y_i$) e a velocidade inicial ($v_i$).

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
time = np.linspace(0.0, 5.0, 50)
y0 = float(input("Digite a altura inicial:"))
v0 = float(input("Digite a velocidade inicial:"))
posicao = y0 + v0*time - (4.9 * time ** 2)
```

```python
plt.plot(time, posicao, 'r:')
plt.axis((0, 5, 0, y0+v0*2))
plt.title('Posição vs Tempo')
plt.xlabel('Tempo [s]')
plt.ylabel('Posição [m]')
plt.grid(True, linewidth=0.4)
plt.annotate('altura máxima', xy=(v0/9.8,y0+(v0**2)/(19.6)), xytext=((v0/9.8)+1,y0+(v0**2)/(19.6)+3), arrowprops=dict(arrowstyle='->'))
plt.show()
```
